### Addition/Deletion Analysis

```python
import os
from itertools import cycle
from copy import copy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse, Circle
import matplotlib.gridspec as gridspec
import cufflinks as cf
import plotly.offline

from IPython.display import display, HTML
from ipywidgets import interactive_output, HBox, VBox, Layout 
from ipywidgets import Dropdown, Checkbox, ToggleButtons, SelectionRangeSlider, SelectionSlider
import ipywidgets as widgets

cf.go_offline()
pd.set_option('display.max_columns', 100)

# in_dir = '../../output/update/csv/'
in_dir = '../../'
main_fp = os.path.join(in_dir, 'results.csv')
retrain_fp = os.path.join(in_dir, 'retrain.csv')

main_df = pd.read_csv(main_fp)
retrain_df = pd.read_csv(retrain_fp)

main_df

main_df[main_df['dataset'] == 'surgical']
```

```python
def f(dataset, sub_size, log_scale, annotation, metric, criterion, operation,
      exact, cedar):

    n_trees = dataset_dict[dataset][1]
    depth = dataset_dict[dataset][2]

    df = main_df[main_df['dataset'] == dataset]
    df = df[df['operation'] == operation]
    df = df[df['criterion'] == criterion]
    df = df[df['subsample_size'] == sub_size]
    exp_df = df[df['n_estimators'] == n_trees]
    exp_df = exp_df[exp_df['max_depth'] == depth]
    
    exact_df = exp_df[exp_df['model'] == 'exact']
    dart_df = exp_df[exp_df['model'] == 'dart']
    cedar_df = exp_df[(exp_df['model'] == 'cedar')]

    tol_list = ['0.1%', '0.25%', '0.5%', '1.0%']
    shape_list = ['o', '^', 's', 'd']
    shape_size = 10
    
    cedar_colors = np.random.rand(8, 3)

    fig, axs = plt.subplots(1, 3, figsize=(15, 5))

    # plot efficiency
    ax = axs[0]
    ax.set_xlabel('topd')
    ax.set_ylabel('speedup vs naive (x)')
    ax.set_title('Efficiency')
    ax.errorbar(dart_df['topd'], dart_df['n_model'], yerr=dart_df['n_model_std'], label='dart')
    if annotation:
        for tol, topd, shape in zip(tol_list, dataset_dict[dataset][3], shape_list):
            if topd == 0:
                continue
            x = dart_df['topd'].iloc[topd - 1]
            y = dart_df['n_model'].iloc[topd - 1]
            ax.plot(x, y, 'k{}'.format(shape), label='tol={}'.format(tol), ms=shape_size)
    if exact:
        ax.axhline(exact_df['n_model'].values[0], color='k', linestyle='--', label='exact')
    if cedar:
        for i in range(len(cedar_df)):
            n_model = cedar_df.iloc[i]['n_model']
            epsilon = cedar_df.iloc[i]['epsilon']
            lmbda = cedar_df.iloc[i]['lmbda']
            ax.axhline(n_model, linestyle=':', color=cedar_colors[i],
                       label=r'$\epsilon={}$, $\lambda={}$'.format(epsilon, lmbda))
    if log_scale:
        ax.set_yscale('log')
    ax.legend()

    # plot utility
    ax = axs[1]
    ax.set_xlabel('topd')
    ax.set_ylabel('error increase ({})'.format(metric))
    ax.set_title('Utility')
    ax.errorbar(dart_df['topd'], dart_df['{}_diff_mean'.format(metric)], yerr=dart_df['{}_diff_std'.format(metric)])
    if annotation:
        for tol, topd, shape in zip(tol_list, dataset_dict[dataset][3], shape_list):
            if topd == 0:
                continue
            x = dart_df['topd'].iloc[topd - 1]
            y = dart_df['{}_diff_mean'.format(metric)].iloc[topd - 1]
            ax.plot(x, y, 'k{}'.format(shape), label='tol={}'.format(tol), ms=shape_size)
    if cedar:
        for i in range(len(cedar_df)):
            metric_val = cedar_df.iloc[i]['{}_diff_mean'.format(metric)]
            epsilon = cedar_df.iloc[i]['epsilon']
            lmbda = cedar_df.iloc[i]['lmbda']
            ax.axhline(metric_val, linestyle=':', color=cedar_colors[i],
                       label=r'$\epsilon={}$, $\lambda={}$'.format(epsilon, lmbda))
    if exact:
        ax.axhline(0, color='k', linestyle='--')

    # plot retrains
    ax = axs[2]
    ax.set_xlabel('depth')
    ax.set_ylabel('no. retrains')
    ax.set_title('Retrain Depths')
    for i, row in enumerate(dart_df.itertuples(index=False)):
        linestyle = '-' if i < 10 else '--'
        temp = retrain_df[retrain_df['id'] == row.id]
        retrains = temp.iloc[0].values[1:]
        depths = np.arange(retrains.shape[0])
        ax.plot(depths, retrains, linestyle=linestyle, label='top {}'.format(row.topd))
    if exact:
        temp = retrain_df[retrain_df['id'].isin(exact_df['id'])]
        retrains = temp.iloc[0].values[1:]
        depths = np.arange(retrains.shape[0])
        ax.plot(depths, retrains, 'k--')
    ax.legend(ncol=2)

    plt.tight_layout()
    plt.show()

# UI
dataset = Dropdown(description='Dataset', options=['surgical', 'vaccine', 'adult',
                                                   'bank_marketing', 'flight_delays', 'diabetes',
                                                   'olympics',
                                                   'census', 'credit_card',
                                                   'synthetic', 'higgs'])
log_scale = Checkbox(description='Log scale', value=True)
annotation = Checkbox(description='annotation', value=True)
metric = SelectionSlider(description='Metric', options=['auc', 'acc', 'ap'], value='acc')
criterion = SelectionSlider(description='Criterion', options=['gini', 'entropy'], value='gini')
operation = SelectionSlider(description='Operation', options=['delete', 'add'], value='delete')
sub_size = SelectionSlider(description='Sample size', options=[1, 100], value=1)
cedar = Checkbox(description='cedar', value=False)
exact = Checkbox(description='exact', value=True)

# create ui
box_1 = VBox([dataset, metric, criterion, operation])
box_2 = VBox([sub_size, log_scale, annotation])
box_3 = VBox([exact, cedar])
ui = HBox([box_1, box_2, box_3])

dataset_dict = {'surgical': ('acc', 250, 20, [2, 13, 18, 19]),
                'vaccine': ('acc', 250, 10, [1, 4, 5, 7]),
                'adult': ('acc', 10, 20, [15, 17, 18, 19]),
                'bank_marketing': ('auc', 100, 10, [9, 9, 9, 9]),
                'flight_delays': ('auc', 250, 20, [4, 9, 11, 15]),
                'diabetes': ('acc', 250, 20, [6, 11, 15, 19]),
                'olympics': ('auc', 250, 20, [0, 1, 2, 4]),
                'census': ('auc', 250, 20, [9, 13, 15, 18]),
                'credit_card': ('ap', 250, 20, [5, 5, 10, 12]),
                'synthetic': ('acc', 250, 20, [2, 4, 6, 9]),
                'higgs': ('acc', 100, 10, [1, 2, 4, 5])}

def update_parameters(*args):
    metric.value = dataset_dict[dataset.value][0]

dataset.observe(update_parameters, names='value')

out = interactive_output(f, {'dataset': dataset, 'criterion': criterion, 'operation': operation,
                             'sub_size': sub_size, 'log_scale': log_scale, 'metric': metric,
                             'exact': exact, 'cedar': cedar, 'annotation': annotation})
display(ui, out)
```
