### Amortized Experiments 2

```python
import os
from itertools import cycle
from copy import copy

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import cufflinks as cf
import plotly.offline

from IPython.display import display, HTML
from ipywidgets import interactive_output, HBox, VBox, Layout 
from ipywidgets import Dropdown, Checkbox, ToggleButtons, SelectionRangeSlider, SelectionSlider
import ipywidgets as widgets

cf.go_offline()
pd.set_option('display.max_columns', 100)

experiment = 'deletion'

data_dir = '../../'
df = pd.read_csv(os.path.join(data_dir, 'results.csv'))
df
```

```python
def f(dataset, metric, adversary, naive, cedar, exact, exact_alternates):
    
    lines = ['--', '-.', ':']
    colors = ['r', 'g', 'purple', 'orange', 'blue']
    linecolors = [(c, l) for c in colors for l in lines]
    lc_cycler = cycle(linecolors)

    fig, axs = plt.subplots(1, 2, figsize=(15, 5))
    axs = axs.flatten()

    ax = axs[0]
    ax.set_title('Efficiency')
    ax.set_xlabel('top d')
    ax.set_ylabel('amortized (s)')

    ax = axs[1]
    ax.set_title('Utility')
    ax.set_xlabel('top d')
    ax.set_ylabel(metric)
    
    # settings that apply to all methods
    main_df = df[df['adversary'] == adversary]

    # experiment hyperparameters
    temp = df[df['dataset'] == dataset]
    temp = temp[temp['model_type'] == 'dart']
    temp = temp[temp['method'] == 'naive']
    n_trees = copy(temp.iloc[0]['n_estimators'])
    max_depth = copy(temp.iloc[0]['max_depth'])

    # plot naive
    if naive:
        temp = main_df[main_df['dataset'] == dataset]
        temp = temp[temp['model_type'] == 'dart']
        naive_df = temp[temp['method'] == 'naive']
        axs[0].axhline(naive_df.iloc[0]['amortized'], label='naive', color='brown', linestyle='-')
    
    # plot Node cedar type
    if cedar:
        temp = main_df[main_df['dataset'] == dataset]
        temp = temp[temp['model_type'] == 'cedar']
        temp = temp[temp['method'] == 'cedar']
        temp = temp[temp['max_depth'] == max_depth]

        for (epsilon, lmbda, topd), gf in temp.groupby(['epsilon', 'lmbda', 'topd']):
            color, linestyle = next(lc_cycler)
            label = r'topd={}, $\epsilon$={}, $\lambda$={}'.format(topd, epsilon, lmbda)
            axs[0].axhline(gf['amortized'].mean(), color=color, linestyle=linestyle, label=label)
            axs[1].axhline(gf[metric].mean(), color=color, linestyle=linestyle)
        
    # plot DART
    temp = main_df[main_df['dataset'] == dataset]
    temp = temp[temp['model_type'] == 'dart']
    temp = temp[temp['n_estimators'] == n_trees]
    temp = temp[temp['max_depth'] == max_depth]

    results = []
    for (n_estimators, max_depth, topd), gf in temp.groupby(['n_estimators', 'max_depth', 'topd']):
        amortized_mean = gf['amortized'].mean()
        amortized_std = gf['amortized'].std()

        results.append({'topd': topd,
                        'amortized_avg': amortized_mean, 'amortized_std': amortized_std,
                        '{}_avg'.format(metric): gf[metric].mean(), '{}_std'.format(metric): gf[metric].std()})

        if naive and topd in dataset_dict[dataset][1]:
            gf['speedup_vs_naive'] = naive_df['amortized'].values[0] / gf['amortized']
            speedup_mean = gf['speedup_vs_naive'].mean()
            speedup_std = gf['speedup_vs_naive'].std()
            metric_diff = gf['{}_diff_avg'.format(metric)].mean()
            metric_std = gf['{}_diff_avg'.format(metric)].std()

            s = '[DART] topd: {}, amortized: {:.5f}s +/- {:.5f}s, speedup_vs_naive: {:.1f} +/- {:.1f}'
            s += ', {}_difference: {:.3f} +/ {:.3f}'
            print(s.format(topd, amortized_mean, amortized_std, speedup_mean, speedup_std,
                           metric, metric_diff, metric_std))

    df2 = pd.DataFrame(results)
    axs[0].errorbar(df2['topd'], df2['amortized_avg'], yerr=df2['amortized_std'], label='top_rand')
    axs[1].errorbar(df2['topd'], df2['{}_avg'.format(metric)], yerr=df2['{}_std'.format(metric)], label='top_rand')
    
    # plot exact
    if exact:
        temp = main_df[main_df['dataset'] == dataset]
        temp = temp[temp['model_type'] == 'dart']
        temp = temp[temp['method'] == 'exact']
        temp = temp[temp['n_estimators'] == n_trees]
        exact_df = temp[temp['max_depth'] == max_depth]

        axs[0].axhline(exact_df.iloc[0]['amortized'], label='exact', color='black', linestyle='--')
        axs[1].axhline(exact_df.iloc[0][metric], color='black', linestyle='--')
        
        if naive:
            exact_df['speedup_vs_naive'] = naive_df['amortized'].values[0] / exact_df['amortized']
            speedup_mean = exact_df['speedup_vs_naive'].mean()
            speedup_std = exact_df['speedup_vs_naive'].std()
            amortized_mean = exact_df['amortized'].mean()
            amortized_std = exact_df['amortized'].std()

            s = '[EXACT] amortized: {:.5f}s +/- {:.5f}s, speedup_vs_naive: {:.1f} +/- {:.1f}'
            print(s.format(amortized_mean, amortized_std, speedup_mean, speedup_std))
    
    if exact_alternates:
        temp = main_df[main_df['dataset'] == dataset]
        temp = temp[temp['model_type'] == 'dart']
        temp = temp[temp['method'] == 'exact']
        temp = temp[temp['n_estimators'] != n_trees]
        temp = temp[temp['max_depth'] != max_depth]

        for (n_estimators, depth), gf in temp.sort_values('amortized').groupby(['n_estimators', 'max_depth']):
            color, linestyle = next(lc_cycler)
            label = 'exact: t{}_d{}'.format(n_estimators, depth)
            axs[0].axhline(gf.iloc[0]['amortized'], label=label, color=color, linestyle=linestyle)
            axs[1].axhline(gf.iloc[0][metric], color=color, linestyle=linestyle)

    # plot settings
    title = 'Trees: {}, Depth: {}'.format(n_trees, max_depth)
    axs[0].legend(bbox_to_anchor=(2.8, 1.01), title=title)
    axs[0].set_yscale('log')

    plt.show()
    
# UI
dataset = Dropdown(description='Dataset', options=['surgical', 'vaccine', 'adult',
                                                   'bank_marketing', 'flight_delays', 'diabetes',
                                                   'olympics',
                                                   'census', 'credit_card',
                                                   'synthetic', 'higgs'])
metric = SelectionSlider(description='Metric', options=['auc', 'acc', 'bacc', 'ap'], value='acc')
adversary = SelectionSlider(description='Adversary', options=['random', 'root', 'worst'], value='random')
naive = Checkbox(description='naive', value=False)
cedar = Checkbox(description='cedar', value=False)
exact = Checkbox(description='exact', value=False)
exact_alternates = Checkbox(description='exact (alternates)', value=False)

# create ui
box_1 = VBox([dataset, metric, adversary])
box_2 = VBox([naive, cedar])
box_3 = VBox([exact, exact_alternates])
ui = HBox([box_1, box_2, box_3])

# reactions
dataset_dict = {'surgical': ('acc', [17, 19]), 'vaccine': ('acc', [6, 7]), 'adult': ('acc', [17, 18]),
                'bank_marketing': ('auc', [9]), 'flight_delays': ('auc', [11, 14]), 'diabetes': ('acc', [16, 19]),
                'olympics': ('auc', [2, 4]), 'census': ('auc', [15, 18]), 'credit_card': ('ap', [10, 12]),
                'synthetic': ('acc', [6, 8]), 'higgs': ('acc', [4, 6])}

def update_metric(*args):
    metric.value = dataset_dict[dataset.value][0]

dataset.observe(update_metric, names='value')

# disply UI
out = interactive_output(f, {'dataset': dataset, 'metric': metric, 'adversary': adversary, 'naive': naive,
                             'cedar': cedar, 'exact': exact, 'exact_alternates': exact_alternates})
display(ui, out)
```
